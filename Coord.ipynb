```python
new_list = [(re.sub('([A-Z])', r' \1', i).strip() + ', Australia') for i in lista]
```

```python
new_list
```

```python
coord = {}
erros = {}
for item in new_list:
    geolocator = Nominatim(user_agent = 'x')
    location = geolocator.geocode(item, timeout = None)
    if location == None:
        erros.update({item : location})
    else:
        coord.update({item : (location.latitude , location.longitude)})
coord.update({'PearceRAAF, Australia' : (-31.8806408, 115.7298859)})
```

```python
df_coord = pd.DataFrame(coord)

for item in coord.keys():
    df_coord.rename(columns = {item: item.replace(", Australia","").replace(" ","")},inplace=True)
                    
df_coord = df_coord.T.rename(columns={0:'latitude',1:'longitude'})
```
